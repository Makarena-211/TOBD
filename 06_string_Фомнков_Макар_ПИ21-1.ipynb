````python
pattern = r'^(?:this[a-zA-Z0-9_\s]*?but)'   # ^ = поиск в начале строки, *? = минимальное совпадение,  

pattern1 = re.compile(pattern)
k = 0
l = 0
ex = []

for i in list(datap['description']):
  if len(re.findall(pattern1, str(i))) == 0:
    k+=1
  if len(re.findall(pattern1, str(i))) != 0:
    l+=1
    ex.append(i)
print(ex[0], ex[1], ex[2], sep = '\n')
   
print(f"Не по шаблону: {k}")
print(f"По шаблону: {l}")
````
