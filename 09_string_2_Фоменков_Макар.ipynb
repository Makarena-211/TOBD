2.2. Удалите стоп-слова из описаний рецептов. Какую долю об общего количества слов составляли стоп-слова? Сравните топ-10 самых часто употребляемых слов до и после удаления стоп-слов.

```python
file = pd.read_csv('preprocessed_descriptions.csv')
recipe_description = ' '.join([str(x) for x in list(file['preprocessed_descriptions'])]).split()
filtered_recipe_description = [word for word in recipe_description if word not in stopwords.words('english')]  #удаление стоп слов
#стоп слова, слова не несущие особой смысловой нагрузки, (в русском их 151)
```

```python
counter1 = Counter(recipe_description).most_common(10)   # от наиболее распространенных до наименее (10), это кол во 
counter2 = Counter(filtered_recipe_description).most_common(10)
print(f'Стоп-слова составляли {(len(recipe_description)-len(filtered_recipe_description))*100/len(recipe_description)}% от общего количества слов.')
print(f'Топ-10 самых часто употребляемых слов до и после удаления стоп-слов.')
print(f'-------------------------------------------------------------------------')
print(f'|{"до":^35}|{"после":^35}|')
print(f'-------------------------------------------------------------------------')
for i in range(10):
    print(f'|{counter1[i][0] : ^35}|{counter2[i][0] : ^35}|')
print(f'-------------------------------------------------------------------------')
```

```python
preprocessed_descriptions = pd.read_csv("preprocessed_descriptions.csv")
data = preprocessed_descriptions.sample(5)
data
```

```python
vectorizer = TfidfVectorizer(analyzer="word", stop_words="english")    #Преобразовать текст в частотные векторы слова с помощью
vectorizer.fit(data["preprocessed_descriptions"])

def vectorizer_processing(x):
    sents = [x["preprocessed_descriptions"]]
    vector = vectorizer.transform(sents) 
    return vector.toarray()

data['TfidfVectorizer'] = data.apply(lambda x: vectorizer_processing(x), axis=1)
```

```python
valuess = [(k,v) for (k, v) in dict(vectorizer.vocabulary_).items()]
valuess.sort(key=lambda x: x[1])
valuess
```
